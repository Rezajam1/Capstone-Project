# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

query = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(query, conn)

plt.hist(df['ConvertedComp'].dropna(), bins=30, color='blue', edgecolor='black')

plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import seaborn as sns

query_age = """
SELECT Age
FROM master
"""

df_age = pd.read_sql_query(query_age, conn)

plt.figure(figsize=(10, 6))
sns.boxplot(x=df_age['Age'].dropna(), color='skyblue')

plt.xlabel('Age')
plt.title('Box Plot of Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
query_age_work_hours = """
SELECT Age, WorkWeekHrs
FROM master
WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL
"""

df_age_work_hours = pd.read_sql_query(query_age_work_hours, conn)

plt.figure(figsize=(10, 6))
plt.scatter(df_age_work_hours['Age'], df_age_work_hours['WorkWeekHrs'], color='green', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age and WorkWeekHrs')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
query_work_code_age = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL AND Age IS NOT NULL
"""

df_work_code_age = pd.read_sql_query(query_work_code_age, conn)

plt.figure(figsize=(12, 8))
plt.scatter(
    df_work_code_age['WorkWeekHrs'],
    df_work_code_age['CodeRevHrs'],
    s=df_work_code_age['Age'] * 5,  # Multiply by a factor to adjust bubble size
    alpha=0.6,
    cmap='viridis',  
)

# Add labels and title
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs (Bubble size represents Age)')

plt.colorbar(label='Age')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
query_top_databases = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df_top_databases = pd.read_sql_query(query_top_databases, conn)

plt.figure(figsize=(8, 8))
plt.pie(df_top_databases['count'], labels=df_top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)

plt.title('Top 5 Databases Respondents Wish to Learn Next Year')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
query_median_hours = """
SELECT Age, 
       AVG(WorkWeekHrs) AS MedianWorkWeekHrs, 
       AVG(CodeRevHrs) AS MedianCodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
ORDER BY Age
"""

df_median_hours = pd.read_sql_query(query_median_hours, conn)

plt.figure(figsize=(10, 6))
plt.bar(df_median_hours['Age'], df_median_hours['MedianWorkWeekHrs'], label='WorkWeekHrs', color='skyblue')
plt.bar(df_median_hours['Age'], df_median_hours['MedianCodeRevHrs'], bottom=df_median_hours['MedianWorkWeekHrs'], label='CodeRevHrs', color='orange')

plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
query_median_comp = """
SELECT Age, 
       AVG(ConvertedComp) AS MedianConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
ORDER BY Age
"""

df_median_comp = pd.read_sql_query(query_median_comp, conn)

plt.figure(figsize=(10, 6))
plt.plot(df_median_comp['Age'], df_median_comp['MedianConvertedComp'], marker='o', color='green', linestyle='-', linewidth=2)

plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.title('Median Converted Compensation for Ages 45 to 60')

plt.grid(True)
plt.show()
```
